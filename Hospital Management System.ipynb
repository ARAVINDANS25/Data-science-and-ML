# Hospital Management System Using Python and SQL

```python
from tkinter import *
from tkinter import ttk  # Import ttk module for combobox
import mysql.connector
from tkinter import messagebox

class Hospital:
    def __init__(self, root):
        self.root = root
        self.root.title("Hospital Management System")
        self.root.geometry("1540x800+0+0")

        self.Nameoftablets = StringVar()
        self.ref = StringVar()
        self.dose = StringVar()
        self.numberoftablet = StringVar()
        self.lot = StringVar()
        self.issuedate = StringVar()
        self.expdate = StringVar()
        self.dailydose = StringVar()
        self.sideeffect = StringVar()
        self.furtherinfo = StringVar()
        self.bloodpressure = StringVar()
        self.storageadvice = StringVar()
        self.medication = StringVar()
        self.nhsnumber = StringVar()
        self.patientname = StringVar()
        self.dob = StringVar()
        self.patientaddress = StringVar()

        lbltitle = Label(self.root, bd=20, relief=RIDGE, text="HOSPITAL MANAGEMENT SYSTEM",
                         fg='red', bg='white', font=("times new roman", 50, "bold"))
        lbltitle.pack(side=TOP, fill=X)

        Dataframe = Frame(self.root, bd=20, relief=RIDGE)
        Dataframe.place(x=0, y=130, width=1530, height=400)

        # Patient Information Frame
        dataframeLeft = LabelFrame(Dataframe, bd=10, relief=RIDGE, padx=10,
                                  font=("arial", 12, "bold"), text="Patient Information")
        dataframeLeft.place(x=0, y=5, width=980, height=350)

        # Prescription Frame
        dataframeRight = LabelFrame(Dataframe, bd=10, relief=RIDGE, padx=10,
                                   font=("arial", 12, "bold"), text="Prescription")
        dataframeRight.place(x=990, y=5, width=460, height=350)

        # Button Frame
        buttonFrame = Frame(self.root, bd=20, relief=RIDGE)
        buttonFrame.place(x=0, y=530, width=1530, height=70)

        # Details Frame
        detailsFrame = Frame(self.root, bd=20, relief=RIDGE)
        detailsFrame.place(x=0, y=600, width=1530, height=190)

        # Labels and Entry fields for Patient Information
        labels = ["Name of Tablet:", "Reference No:", "Dose:", "No of Tablet:", "Lot:", "Issue Date:", "Exp Date:", "Daily Dose:", "Side Effect:"]
        var_list = [self.Nameoftablets, self.ref, self.dose, self.numberoftablet, self.lot, self.issuedate, self.expdate, self.dailydose, self.sideeffect]
        for i, label_text in enumerate(labels):
            label = Label(dataframeLeft, text=label_text, font=("times new roman", 12, "bold"), padx=2, pady=6)
            label.grid(row=i, column=0, sticky=E)  # Align to the right

            entry = Entry(dataframeLeft, font=("arial", 12), width=33, textvariable=var_list[i])
            entry.grid(row=i, column=1, sticky=W)  # Align to the left

        # Additional labels and entry fields for Patient Information
        additional_labels = ["Futher Information:", "Blood Pressure:", "Storage Advice:", "Medication:", "Patient Id NHS Number:", "Patient Name:", "DOB:", "Patient Address:"]
        var_list_additional = [self.furtherinfo, self.bloodpressure, self.storageadvice, self.medication, self.nhsnumber, self.patientname, self.dob, self.patientaddress]
        for i, label_text in enumerate(additional_labels):
            label = Label(dataframeLeft, text=label_text, font=("times new roman", 12, "bold"), padx=2, pady=6)
            label.grid(row=i, column=2, sticky=E)  # Align to the right

            entry = Entry(dataframeLeft, font=("arial", 12), width=33, textvariable=var_list_additional[i])
            entry.grid(row=i, column=3, sticky=W)  # Align to the left

        # Text and Button for Prescription
        self.txtPrescription = Text(dataframeRight, font=("arial", 12, "bold"), width=45, height=16, padx=2, pady=6)
        self.txtPrescription.insert(END, "Write prescription here...")
        self.txtPrescription.grid(row=0, column=0)

        # Additional buttons in Prescription Frame
        prescription_buttons = ["Prescription", "Prescription Data", "Update", "Delete", "Clear", "Exit"]
        for i, button_text in enumerate(prescription_buttons):
            btn = Button(buttonFrame, text=button_text, bg="green", fg="white", font=("arial", 12, "bold"), width=23, padx=2, pady=6)
            btn.grid(row=0, column=i+1)

            # Associate the iPrescriptionData method with the "Prescription Data" button
            if button_text == "Prescription Data":
                btn["command"] = self.iPrescriptionData

            # Corrected the button_text for "Update" and associated it with the update_data method
            if button_text == "Update":
                btn["command"] = self.update_data
                
            if button_text == "Prescription":
                btn["command"] = self.iPrescription
                
            if button_text == "Delete":
                btn["command"] = self.idelete
            
            if button_text == 'Clear':
                btn["command"] = self.clear
                
            if button_text == 'Exit':
                btn["command"] = self.iexit

                

        scroll_x = ttk.Scrollbar(detailsFrame, orient=HORIZONTAL)
        scroll_y = ttk.Scrollbar(detailsFrame, orient=VERTICAL)

        self.hospital_table = ttk.Treeview(detailsFrame, column=("Name of Tablet", "Reference No", "Dose", "No of Tablet",
                                                         "Lot", "Issue Date", "Exp Date", "Daily Dose", "Side Effect",  # Add missing comma
                                                         "Futher Information", "Blood Pressure",
                                                         "Storage Advice", "Medication", "NHS Number", "Patient Name",
                                                         "DOB", "Patient Address"),
                                  xscrollcommand=scroll_x.set, yscrollcommand=scroll_y.set)

        scroll_x.pack(side=BOTTOM, fill=X)
        scroll_y.pack(side=RIGHT, fill=Y)

        scroll_x.config(command=self.hospital_table.xview)
        scroll_y.config(command=self.hospital_table.yview)
        self.hospital_table.heading("Name of Tablet", text="Name of Tablet")
        self.hospital_table.heading("Reference No", text="Reference No")
        self.hospital_table.heading("Dose", text="Dose")
        self.hospital_table.heading("No of Tablet", text="No of Tablet")
        self.hospital_table.heading("Lot", text="Lot")
        self.hospital_table.heading("Issue Date", text="Issue Date")
        self.hospital_table.heading("Exp Date", text="Exp Date")
        self.hospital_table.heading("Daily Dose", text="Daily Dose")
        self.hospital_table.heading("Side Effect", text="Side Effect")  # Add missing heading
        self.hospital_table.heading("Futher Information", text="Futher Information")  # Corrected spelling
        self.hospital_table.heading("Blood Pressure", text="Blood Pressure")  # Added missing heading
        self.hospital_table.heading("Storage Advice", text="Storage Advice")
        self.hospital_table.heading("Medication", text="Medication")  # Added missing heading
        self.hospital_table.heading("NHS Number", text="NHS Number")
        self.hospital_table.heading("Patient Name", text="Patient Name")
        self.hospital_table.heading("DOB", text="DOB")
        self.hospital_table.heading("Patient Address", text="Patient Address")

        self.hospital_table["show"] = "headings"

        common_width = 100
        for col in self.hospital_table["columns"]:
            self.hospital_table.column(col, width=common_width, anchor=CENTER)

        self.hospital_table.pack(fill=BOTH, expand=1)
        
        self.hospital_table.bind("<ButtonRelease-1>", self.get_cursor)
        
        self.fetch_data()
        
    def iPrescriptionData(self):
        if self.Nameoftablets.get() == "" or self.ref.get() == "":
            messagebox.showerror("Error", "All fields are required")
        else:
            try:
                conn = mysql.connector.connect(host='localhost', user='root', 
                                       password='thedubbelH@25', database='aravindan_s_db')
                my_cursor = conn.cursor()

                # Corrected typo in execute method, and used placeholders in the SQL query
                my_cursor.execute("INSERT INTO hospital (Nameoftablets, ref, dose, numberoftablet, lot, issuedate, expdate, "
                                  "dailydose, sideeffect, furtherinfo, bloodpressure, storageadvice, medication, nhsnumber, "
                                  "patientname, dob, patientaddress) "
                                  "VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)",
                                  (self.Nameoftablets.get(), self.ref.get(), self.dose.get(), self.numberoftablet.get(),
                                   self.lot.get(), self.issuedate.get(), self.expdate.get(), self.dailydose.get(),
                                   self.sideeffect.get(), self.furtherinfo.get(), self.bloodpressure.get(),
                                   self.storageadvice.get(), self.medication.get(), self.nhsnumber.get(),
                                   self.patientname.get(), self.dob.get(), self.patientaddress.get()))

                conn.commit()
                self.fetch_data()
                conn.close()
                messagebox.showinfo("Success", "Data added successfully")
            except Exception as e:
                messagebox.showerror("Error", f"Error in adding data: {str(e)}")
                
    def update_data(self):
        try:
            conn = mysql.connector.connect(host='localhost', user='root', 
                                       password='thedubbelH@25', database='aravindan_s_db')
            my_cursor = conn.cursor()

            my_cursor.execute("UPDATE hospital SET Nameoftablets=%s, ref=%s, dose=%s, numberoftablet=%s, lot=%s, issuedate=%s, expdate=%s, "
                              "dailydose=%s, sideeffect=%s, furtherinfo=%s, bloodpressure=%s, storageadvice=%s, medication=%s, nhsnumber=%s, "
                              "patientname=%s, dob=%s, patientaddress=%s WHERE ref=%s",
                              (self.Nameoftablets.get(), self.ref.get(), self.dose.get(), self.numberoftablet.get(),
                               self.lot.get(), self.issuedate.get(), self.expdate.get(), self.dailydose.get(),
                               self.sideeffect.get(), self.furtherinfo.get(), self.bloodpressure.get(),
                               self.storageadvice.get(), self.medication.get(), self.nhsnumber.get(),
                               self.patientname.get(), self.dob.get(), self.patientaddress.get(), self.ref.get()))

            conn.commit()
            self.fetch_data()  # Update the displayed data after the update
            conn.close()
            messagebox.showinfo("Success", "Data updated successfully")
        except Exception as e:
            messagebox.showerror("Error", f"Error in updating data: {str(e)}")

    def fetch_data(self):
        try:
            conn = mysql.connector.connect(host='localhost', user='root', 
                                       password='thedubbelH@25', database='aravindan_s_db')
            my_cursor = conn.cursor()
            my_cursor.execute("select * from hospital")

            rows = my_cursor.fetchall()
            if len(rows) != 0:
                self.hospital_table.delete(*self.hospital_table.get_children())
                for i in rows:
                    self.hospital_table.insert("", END, values=i)
            conn.close()
        except Exception as e:
            messagebox.showerror("Error", f"Error in fetching data: {str(e)}")

    def get_cursor(self, event=""):
        try:
            cursor_row = self.hospital_table.focus()
            content = self.hospital_table.item(cursor_row)
            row = content["values"]
            self.Nameoftablets.set(row[0])
            self.ref.set(row[1])
            self.dose.set(row[2])
            self.numberoftablet.set(row[3])
            self.lot.set(row[4])
            self.issuedate.set(row[5])
            self.expdate.set(row[6])
            self.dailydose.set(row[7])
            self.sideeffect.set(row[8])
            self.furtherinfo.set(row[9])
            self.bloodpressure.set(row[10])
            self.storageadvice.set(row[11])
            self.medication.set(row[12])
            self.nhsnumber.set(row[13])
            self.patientname.set(row[14])
            self.dob.set(row[15])
            self.patientaddress.set(row[16])
        except Exception as e:
            messagebox.showerror("Error", f"Error in getting cursor: {str(e)}")
        
    
    def iPrescription(self):
        self.txtPrescription.delete("1.0", END)  # Clear previous content

        self.txtPrescription.insert(END, "Name of Tablets:\t\t\t" + self.Nameoftablets.get() + "\n")
        self.txtPrescription.insert(END, "Reference No:\t\t\t" + self.ref.get() + "\n")
        self.txtPrescription.insert(END, "Dose:\t\t\t" + self.dose.get() + "\n")
        self.txtPrescription.insert(END, "No of Tablet:\t\t\t" + self.numberoftablet.get() + "\n")
        self.txtPrescription.insert(END, "Lot:\t\t\t" + self.lot.get() + "\n")
        self.txtPrescription.insert(END, "Issue Date:\t\t\t" + self.issuedate.get() + "\n")
        self.txtPrescription.insert(END, "Exp Date:\t\t\t" + self.expdate.get() + "\n")
        self.txtPrescription.insert(END, "Daily Dose:\t\t\t" + self.dailydose.get() + "\n")
        self.txtPrescription.insert(END, "Side Effect:\t\t\t" + self.sideeffect.get() + "\n")
        self.txtPrescription.insert(END, "Further Information:\t\t\t" + self.furtherinfo.get() + "\n")
        self.txtPrescription.insert(END, "Blood Pressure:\t\t\t" + self.bloodpressure.get() + "\n")
        self.txtPrescription.insert(END, "Storage Advice:\t\t\t" + self.storageadvice.get() + "\n")
        self.txtPrescription.insert(END, "Medication:\t\t\t" + self.medication.get() + "\n")
        self.txtPrescription.insert(END, "NHS Number:\t\t\t" + self.nhsnumber.get() + "\n")
        self.txtPrescription.insert(END, "Patient Name:\t\t\t" + self.patientname.get() + "\n")
        self.txtPrescription.insert(END, "DOB:\t\t\t" + self.dob.get() + "\n")
        self.txtPrescription.insert(END, "Patient Address:\t\t\t" + self.patientaddress.get() + "\n")
        
    def idelete(self):
            conn = mysql.connector.connect(host='localhost', user='root', 
                                       password='thedubbelH@25', database='aravindan_s_db')
            my_cursor = conn.cursor()
            query = "DELETE FROM hospital WHERE ref=%s"
            value = (self.ref.get(),)
            my_cursor.execute(query, value)
        
            conn.commit()
            conn.close()
            self.fetch_data()  # Corrected the method name
            messagebox.showinfo("Delete", "Patient has been deleted successfully")  # Corrected the message box method
            
    def clear(self):
        self.Nameoftablets.set("")
        self.ref.set("")
        self.dose.set("")
        self.numberoftablet.set("")  # Corrected variable name
        self.lot.set("")
        self.issuedate.set("")
        self.expdate.set("")
        self.dailydose.set("")
        self.sideeffect.set("")
        self.furtherinfo.set("")
        self.bloodpressure.set("")
        self.storageadvice.set("")
        self.medication.set("")
        self.nhsnumber.set("")
        self.patientname.set("")
        self.dob.set("")
        self.patientaddress.set("")
        self.txtPrescription.delete("1.0", END)
    
    def iexit(self):
        iexit = messagebox.askyesno("Hospital Management System", "Confirm you want to exit")
        if iexit > 0:
            root.destroy()
            return

 
        
root = Tk()
ob = Hospital(root)
root.mainloop()
```
