```python
from tkinter import *
from tkinter import ttk
import mysql.connector
from tkinter import messagebox

class LibraryManagementSystem:
    def __init__(self, root):
        self.root = root
        self.root.title("Library Management System")
        self.root.geometry("1540x800+0+0")

        self.book_id = StringVar()
        self.book_title = StringVar()
        self.author = StringVar()
        self.isbn = StringVar()
        self.genre = StringVar()
        self.published_year = StringVar()
        self.quantity = StringVar()
        self.member_type = StringVar()
        self.prn_no = StringVar()
        self.id_no = StringVar()
        self.first_name = StringVar()
        self.last_name = StringVar()
        self.address = StringVar()
        self.post_code = StringVar()
        self.mobile_number = StringVar()
        self.fine = StringVar()

        lbltitle = Label(self.root, bd=20, relief=RIDGE, text="LIBRARY MANAGEMENT SYSTEM",
                         fg='Green', bg='white', font=("times new roman", 50, "bold"))
        lbltitle.pack(side=TOP, fill=X)

        Dataframe = Frame(self.root, bd=20, relief=RIDGE)
        Dataframe.place(x=0, y=130, width=1530, height=400)

        # Book Information Frame
        dataframeLeft = LabelFrame(Dataframe, bd=10, relief=RIDGE, padx=10,
                                   font=("arial", 12, "bold"), text="Book Information")
        dataframeLeft.place(x=0, y=5, width=980, height=350)
        
        dataframeRight = LabelFrame(Dataframe, bd=10, relief=RIDGE, padx=10,
                                   font=("arial", 12, "bold"), text="Book Details")
        dataframeRight.place(x=990, y=5, width=460, height=350)
        


        # Button Frame
        buttonFrame = Frame(self.root, bd=20, relief=RIDGE)
        buttonFrame.place(x=0, y=530, width=1530, height=70)

        # Details Frame
        detailsFrame = Frame(self.root, bd=20, relief=RIDGE)
        detailsFrame.place(x=0, y=600, width=1530, height=190)

        # Labels and Entry fields for Book Information
        labels = ["Member Type:", "PRN No:", "ID No:", "First Name:", "Last Name:", "Address:", "Post Code:",
                  "Mobile Number:"]
        var_list = [self.member_type, self.prn_no, self.id_no, self.first_name, self.last_name,
                    self.address, self.post_code, self.mobile_number]
        
        for i, label_text in enumerate(labels):
            label = Label(dataframeLeft, text=label_text, font=("times new roman", 12, "bold"), padx=2, pady=6)
            label.grid(row=i, column=0, sticky=E)

            entry = Entry(dataframeLeft, font=("arial", 12), width=33, textvariable=var_list[i])
            entry.grid(row=i, column=1, sticky=W)
            
        labels = ["Book ID:", "Book Title:", "Author:", "ISBN:", "Genre:", "Published Year:", "Quantity:","Fine:"]
                  
        var_list = [self.book_id, self.book_title, self.author, self.isbn, self.genre, self.published_year,
                    self.quantity,self.fine]
        for i, label_text in enumerate(labels):
            label = Label(dataframeLeft, text=label_text, font=("times new roman", 12, "bold"), padx=2, pady=6)
            label.grid(row=i, column=2, sticky=E)

            entry = Entry(dataframeLeft, font=("arial", 12), width=33, textvariable=var_list[i])
            entry.grid(row=i, column=3, sticky=W)
            
            
        book_list = ["Book 1", "Book 2", "Book 3", "Book 4", "Book 5", "Book 6", "Book 7", "Book 8", "Book 9", "Book 10"]
        self.selected_book = StringVar()
        self.selected_book.set("Select a book")
        combo_book = ttk.Combobox(dataframeRight, textvariable=self.selected_book, values=book_list, state="readonly")
        combo_book.grid(row=0, column=0)
    
    

        # Button for Book Information
        btnAdd = Button(buttonFrame, text="Add Book", bg="green", fg="white", font=("arial", 12, "bold"), width=23, padx=2,
                        pady=6, command=self.add_book)
        btnAdd.grid(row=0, column=1)

        btnUpdate = Button(buttonFrame, text="Update Book", bg="blue", fg="white", font=("arial", 12, "bold"), width=23, padx=2,
                           pady=6, command=self.update_book)
        btnUpdate.grid(row=0, column=2)

        btnDelete = Button(buttonFrame, text="Delete Book", bg="red", fg="white", font=("arial", 12, "bold"), width=23, padx=2,
                           pady=6, command=self.delete_book)
        btnDelete.grid(row=0, column=3)

        btnClear = Button(buttonFrame, text="Clear", bg="orange", fg="white", font=("arial", 12, "bold"), width=23, padx=2,
                          pady=6, command=self.clear)
        btnClear.grid(row=0, column=4)

        btnExit = Button(buttonFrame, text="Exit", bg="gray", fg="white", font=("arial", 12, "bold"), width=23, padx=2,
                         pady=6, command=self.root.destroy)
        btnExit.grid(row=0, column=5)

        # Details Table
        scroll_x = ttk.Scrollbar(detailsFrame, orient=HORIZONTAL)
        scroll_y = ttk.Scrollbar(detailsFrame, orient=VERTICAL)

        self.library_table = ttk.Treeview(detailsFrame, column=("Book ID", "Book Title", "Author", "ISBN", "Genre",
                                                                "Published Year", "Quantity", "Member Type", "PRN No",
                                                                "ID No", "First Name", "Last Name", "Address", "Post Code",
                                                                "Mobile Number",  "Fine"),
                                          xscrollcommand=scroll_x.set, yscrollcommand=scroll_y.set)

        scroll_x.pack(side=BOTTOM, fill=X)
        scroll_y.pack(side=RIGHT, fill=Y)

        scroll_x.config(command=self.library_table.xview)
        scroll_y.config(command=self.library_table.yview)

        for col in self.library_table["columns"]:
            self.library_table.column(col, width=100, anchor=CENTER)

        self.library_table.heading("Book ID", text="Book ID")
        self.library_table.heading("Book Title", text="Book Title")
        self.library_table.heading("Author", text="Author")
        self.library_table.heading("ISBN", text="ISBN")
        self.library_table.heading("Genre", text="Genre")
        self.library_table.heading("Published Year", text="Published Year")
        self.library_table.heading("Quantity", text="Quantity")
        self.library_table.heading("Member Type", text="Member Type")
        self.library_table.heading("PRN No", text="PRN No")
        self.library_table.heading("ID No", text="ID No")
        self.library_table.heading("First Name", text="First Name")
        self.library_table.heading("Last Name", text="Last Name")
        self.library_table.heading("Address", text="Address")
        self.library_table.heading("Post Code", text="Post Code")
        self.library_table.heading("Mobile Number", text="Mobile Number")
        self.library_table.heading("Fine", text="Fine")

        self.library_table["show"] = "headings"
        self.library_table.pack(fill=BOTH, expand=1)

        self.library_table.bind("<ButtonRelease-1>", self.get_cursor)

        self.fetch_data()

    def add_book(self):
        if any(var.get() == "" for var in [self.book_title, self.author, self.isbn, self.genre, self.published_year,
                                           self.quantity, self.member_type, self.prn_no, self.id_no, self.first_name,
                                           self.last_name, self.address, self.post_code, self.mobile_number,
                                            self.fine]):
            messagebox.showerror("Error", "All fields are required")
        else:
            try:
                conn = mysql.connector.connect(host='localhost', user='root', password='thedubbelH@25',
                                               database='library_management_db')
                my_cursor = conn.cursor()

                my_cursor.execute("INSERT INTO books (book_title, author, isbn, genre, published_year, quantity, "
                                  "member_type, prn_no, id_no, first_name, last_name, address, post_code, "
                                  "mobile_number, fine) "
                                  "VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)",
                                  (self.book_title.get(), self.author.get(), self.isbn.get(), self.genre.get(),
                                   self.published_year.get(), self.quantity.get(), self.member_type.get(),
                                   self.prn_no.get(), self.id_no.get(), self.first_name.get(), self.last_name.get(),
                                   self.address.get(), self.post_code.get(), self.mobile_number.get(),
                                   self.fine.get()))

                conn.commit()
                self.fetch_data()
                conn.close()
                messagebox.showinfo("Success", "Book added successfully")
            except Exception as e:
                messagebox.showerror("Error", f"Error in adding book: {str(e)}")

    def clear(self):
        # Clear entry fields
        for var in [self.book_id, self.book_title, self.author, self.isbn, self.genre, self.published_year,
                    self.quantity, self.member_type, self.prn_no, self.id_no, self.first_name, self.last_name,
                    self.address, self.post_code, self.mobile_number, self.fine]:
            var.set("")

        # Clear treeview
        for record in self.library_table.get_children():
            self.library_table.delete(record)
            
    def update_book(self):
        if self.book_id.get() == "":
            messagebox.showerror("Error", "Please select a record to update")
        else:
            try:
                conn = mysql.connector.connect(host='localhost', user='root', password='thedubbelH@25',
                                               database='library_management_db')
                my_cursor = conn.cursor()

                my_cursor.execute("UPDATE books SET book_title=%s, author=%s, isbn=%s, genre=%s, "
                                  "published_year=%s, quantity=%s, member_type=%s, prn_no=%s, id_no=%s, "
                                  "first_name=%s, last_name=%s, address=%s, post_code=%s, mobile_number=%s, "
                                   "fine=%s WHERE book_id=%s",
                                  (self.book_title.get(), self.author.get(), self.isbn.get(), self.genre.get(),
                                   self.published_year.get(), self.quantity.get(), self.member_type.get(),
                                   self.prn_no.get(), self.id_no.get(), self.first_name.get(), self.last_name.get(),
                                   self.address.get(), self.post_code.get(), self.mobile_number.get(),
                                    self.fine.get(), self.book_id.get()))

                conn.commit()
                self.fetch_data()
                conn.close()
                messagebox.showinfo("Success", "Book updated successfully")
                self.clear()
            except Exception as e:
                messagebox.showerror("Error", f"Error in updating book: {str(e)}")

    def delete_book(self):
        selected_item = self.library_table.selection()
        if not selected_item:
            messagebox.showerror("Error", "Please select a record to delete")
        else:
            confirmation = messagebox.askyesno("Confirmation", "Are you sure you want to delete this record?")
            if confirmation:
                try:
                    conn = mysql.connector.connect(host='localhost', user='root', password='thedubbelH@25',
                                                   database='library_management_db')
                    my_cursor = conn.cursor()

                    my_cursor.execute("DELETE FROM books WHERE book_id=%s", (self.book_id.get(),))

                    conn.commit()
                    self.fetch_data()
                    conn.close()
                    messagebox.showinfo("Success", "Book deleted successfully")
                    self.clear()
                except Exception as e:
                    messagebox.showerror("Error", f"Error in deleting book: {str(e)}")
                    
    
    def get_cursor(self, event=""):
        cursor_row = self.library_table.focus()
        content = self.library_table.item(cursor_row)
        row = content["values"]
        try:
        # Set the values in the respective Entry widgets
            self.book_id.set(row[0])
            self.book_title.set(row[1])
            self.author.set(row[2])
            self.isbn.set(row[3])
            self.genre.set(row[4])
            self.published_year.set(row[5])
            self.quantity.set(row[6])
            self.member_type.set(row[7])
            self.prn_no.set(row[8])
            self.id_no.set(row[9])
            self.first_name.set(row[10])
            self.last_name.set(row[11])
            self.address.set(row[12])
            self.post_code.set(row[13])
            self.mobile_number.set(row[14])
            self.fine.set(row[15])
            
        except Exception as e:

            messagebox.showerror("Error", f"Error in getting cursor: {str(e)}")
        
    def fetch_data(self):
        try:
            conn = mysql.connector.connect(host='localhost', user='root', password='thedubbelH@25',
                                       database='library_management_db')
            my_cursor = conn.cursor()
            my_cursor.execute("SELECT * FROM books")

            rows = my_cursor.fetchall()
            if len(rows) != 0:
                self.library_table.delete(*self.library_table.get_children())
                for i in rows:
                    self.library_table.insert("", END, values=i)
            conn.close()
        except Exception as e:
            messagebox.showerror("Error", f"Error in fetching data: {str(e)}")



    def iexit(self):
        iexit = messagebox.askyesno("Confirm you want to exit")
        if iexit > 0:
            root.destroy()
            return
# Rest of the code remains unchanged

root = Tk()
ob = LibraryManagementSystem(root)
root.mainloop()

```
