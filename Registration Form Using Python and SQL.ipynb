```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import mysql.connector

class RegistrationForm:
    def __init__(self, root):
        self.root = root
        self.root.title("Computer Institute Registration Form")
        self.root.geometry("600x500")
        self.root.resizable(False, False)

        # Establish a connection to the MySQL database
        self.conn = mysql.connector.connect(
            host='localhost',
            user='root',
            password='thedubbelH@25',
            database='aravindan_s_db'
        )

        # Dictionary to store course fees
        self.course_fees = {
            "Python Programming": 200,
            "Web Development": 250,
            "Data Science": 300,
            "Java Programming": 220,
            "Database Management": 200
        }

        self.application_number = StringVar()
        self.student_name = StringVar()
        self.mobile_number = StringVar()
        self.email = StringVar()
        self.address = StringVar()
        self.education = StringVar()
        self.selected_course = StringVar()
        self.course_fee = StringVar()

        # Application Number
        lbl_application_number = Label(root, text="Application Number:", font=("Arial", 12))
        lbl_application_number.grid(row=0, column=0, padx=10, pady=10, sticky=W)

        entry_application_number = Entry(root, textvariable=self.application_number, font=("Arial", 12), width=20)
        entry_application_number.grid(row=0, column=1, padx=10, pady=10, sticky=W)

        # Student Name
        lbl_student_name = Label(root, text="Student Name:", font=("Arial", 12))
        lbl_student_name.grid(row=1, column=0, padx=10, pady=10, sticky=W)

        entry_student_name = Entry(root, textvariable=self.student_name, font=("Arial", 12), width=20)
        entry_student_name.grid(row=1, column=1, padx=10, pady=10, sticky=W)

        # Mobile Number
        lbl_mobile_number = Label(root, text="Mobile Number:", font=("Arial", 12))
        lbl_mobile_number.grid(row=2, column=0, padx=10, pady=10, sticky=W)

        entry_mobile_number = Entry(root, textvariable=self.mobile_number, font=("Arial", 12), width=20)
        entry_mobile_number.grid(row=2, column=1, padx=10, pady=10, sticky=W)

        # Email
        lbl_email = Label(root, text="Email:", font=("Arial", 12))
        lbl_email.grid(row=3, column=0, padx=10, pady=10, sticky=W)

        entry_email = Entry(root, textvariable=self.email, font=("Arial", 12), width=20)
        entry_email.grid(row=3, column=1, padx=10, pady=10, sticky=W)

        # Address
        lbl_address = Label(root, text="Address:", font=("Arial", 12))
        lbl_address.grid(row=4, column=0, padx=10, pady=10, sticky=W)

        entry_address = Entry(root, textvariable=self.address, font=("Arial", 12), width=20)
        entry_address.grid(row=4, column=1, padx=10, pady=10, sticky=W)

        # Education
        lbl_education = Label(root, text="Education:", font=("Arial", 12))
        lbl_education.grid(row=5, column=0, padx=10, pady=10, sticky=W)

        entry_education = Entry(root, textvariable=self.education, font=("Arial", 12), width=20)
        entry_education.grid(row=5, column=1, padx=10, pady=10, sticky=W)

        # Course
        lbl_course = Label(root, text="Select Course:", font=("Arial", 12))
        lbl_course.grid(row=6, column=0, padx=10, pady=10, sticky=W)

        # Combobox for Course selection
        courses = ["Python Programming", "Web Development", "Data Science", "Java Programming", "Database Management"]
        combo_course = ttk.Combobox(root, values=courses, textvariable=self.selected_course, font=("Arial", 12), width=18)
        combo_course.grid(row=6, column=1, padx=10, pady=10, sticky=W)
        combo_course.bind("<<ComboboxSelected>>", self.show_fee)

        # Fee
        lbl_fee = Label(root, text="Course Fee:", font=("Arial", 12))
        lbl_fee.grid(row=7, column=0, padx=10, pady=10, sticky=W)

        entry_fee = Entry(root, textvariable=self.course_fee, font=("Arial", 12), state="readonly", width=20)
        entry_fee.grid(row=7, column=1, padx=10, pady=10, sticky=W)

        # Submit Button
        btn_submit = Button(root, text="Submit", command=self.submit_form, font=("Arial", 12), bg="green", fg="white")
        btn_submit.grid(row=8, column=0, columnspan=2, pady=20)

    def show_fee(self, event):
        selected_course = self.selected_course.get()
        fee = self.course_fees.get(selected_course, 0)
        self.course_fee.set(f"${fee}")

    def submit_form(self):
        try:
            cursor = self.conn.cursor()
            query = "INSERT INTO registrations (application_number, student_name, mobile_number, email, address, education, selected_course, course_fee) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)"
            data = (
                self.application_number.get(),
                self.student_name.get(),
                self.mobile_number.get(),
                self.email.get(),
                self.address.get(),
                self.education.get(),
                self.selected_course.get(),
                self.course_fee.get()
            )
            cursor.execute(query, data)
            self.conn.commit()
            messagebox.showinfo("Success", "Registration successful")
        except Exception as e:
            messagebox.showerror("Error", f"Error in database operation: {str(e)}")
        finally:
            self.conn.close()

if __name__ == "__main__":
    root = Tk()
    app = RegistrationForm(root)
    root.mainloop()
```
